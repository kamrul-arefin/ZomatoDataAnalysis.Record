```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# Sample text to generate word cloud from
comment_word = "Pizza Burger Pasta Sushi Tacos Biryani Pizza Sushi Sushi Pasta"

# Use default stopwords provided by WordCloud
stopwords = set(STOPWORDS)

# Create word cloud
wordcloud = WordCloud(
    width=1500,
    height=1500,
    background_color='white',
    stopwords=stopwords,
    min_font_size=10
).generate(comment_word)

# Plot the WordCloud image
plt.figure(figsize=(10, 10))
plt.title('Restaurants Cuisine - Top Restaurants')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

#top rated resurents are doing well in cuision of
-->North Indian
-->Chinese
-->Italian
-->American
